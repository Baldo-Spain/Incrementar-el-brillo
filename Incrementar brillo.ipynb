# Operaciones aritm√©ticas. 
Multiplicar por un escalar para incrementar brillo

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt 
%matplotlib inline
from IPython.display import Image
```

```python
image = 'malaga.jpg'
img = cv2.imread(image, cv2.IMREAD_COLOR)
print(img.shape)
print(img.dtype)
```

```python
# Recordamos que opencv usa caneles bgr asi que convertimos a RGB
img = img[:,:,::-1]

# y mostramos
plt.imshow(img);
```

```python
# Aplicamos a la imagen original un factor para incrementar o reducir los valores
img_up = cv2.multiply(img, 1.3)
img_down = cv2.multiply(img, 0.7)

fig, axs = plt.subplots(1, 3, figsize=(18, 5))
axs[0].imshow(img_up)
axs[0].set_title("M√°s brillo",fontsize=16) 

axs[1].imshow(img)
axs[1].set_title("Original",fontsize=16)

axs[2].imshow(img_down)
axs[2].set_title("M√°s oscura",fontsize=16);
```
